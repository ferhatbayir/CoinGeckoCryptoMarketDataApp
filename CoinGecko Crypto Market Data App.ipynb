Real-time Bitcoin price pull app

```python

#Ä°mport requests module
import requests

#The adress of the website where you will get the Bitcoin price
url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=eur&order=market_cap_desc&per_page=10&page=1&sparkline=false"

#Send a request to the website and assign the response to the response variable
response = requests.get(url)

#Assign the contents of the response (in JSON format) to the data variable
data = response.json()

#For each dictionary inside the data list
for coin in data:
    #We get the values of the keys I want from the dictionary
    ranking = coin["market_cap_rank"]
    name = coin["name"]
    symbol = coin["symbol"]
    price = coin["current_price"]
    change = coin["price_change_percentage_24h"]

    #Print values to the screen
    print(ranking, name, symbol, "{:.2f}".format(price), "EURO", "{:.2f}%".format(change), sep =" | ", end ="\n\n")




```
